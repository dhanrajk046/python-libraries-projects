Visulations in Pandas

```python
#series and dataframes line chart
import numpy as np
%matplotlib inline

ts = pd.Series(np.random.randn(50), index=pd.date_range('today',periods=50))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(50,4),index=ts.index,
                  columns = ['A','B','X','Y'])
df = df.cumsum()
df.plot()
```
