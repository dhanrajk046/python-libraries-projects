```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup 
import re
```

```python
url = "https://www.hubertiming.com/results/2018MLK" #open link
html = urlopen(url)
```

```python
soup = BeautifulSoup(html)
```

```python
title = soup.title
print(title)
print(title.text)
```

```python
links = soup.find_all('a', href = True)
for link in links:
    print(link.get("href"))
```

```python
data = []
allrows = soup.find_all("tr")
for row in allrows:
    row_list = row.find_all("td")
    dataRow = []
    for cell in row_list:
        dataRow.append(cell.text)
    data.append(dataRow)
titles = data[0]
data = data[4:]
print(data[-2:])
```

```python
df = pd.DataFrame(data)
print(df.head(2))
print(df.tail(2))
```

```python
header_list = []
col_headers = soup.find_all('th')
for col in col_headers:
    header_list.append(col.text)
print(header_list)
```

```python
df.columns = header_list
print(df.head)
```

```python
df.info()
```

```python
df.shape
```

```python
df2 = df.dropna( how ='any')
df2.shape
```

```python
df2['ChipTime_minutes'] = pd.to_timedelta(df2['Chip Time'])
```

```python
print(df2.columns)
```

```python
# Remove unwanted text/spaces
df2['Time'] = df2['Time'].str.strip()

# Convert safely
df2['ChipTime_minutes'] = (
    pd.to_timedelta(df2['Time'], errors='coerce')
    .dt.total_seconds() / 60
)
```

```python
df2['Time'] = df2['Time'].str.strip()

df2['ChipTime_minutes'] = (
    pd.to_timedelta(df2['Time'], errors='coerce')
    .dt.total_seconds() / 60
)

df2[['Gender', 'Time', 'ChipTime_minutes']].head()
```

```python
plt.bar(df2['Gender'], df2['ChipTime_minutes'])
plt.xlabel('Gender')
plt.ylabel('ChipTime_minutes')
plt.title("Comparison of average minutes run by male and female")
```

```python
df2.describe(include=[np.number])
```

```python
df2.boxplot(column='ChipTime_minutes', by='Gender')
plt.ylabel("Run time")
```

```python
plt.scatter(df2['ChipTime_minutes'], df2['Age_i'])
plt.xlabel('ChipTime_minutes')
plt.ylabel('Age')
plt.title('Age vs Chip Time')
plt.show()
```
