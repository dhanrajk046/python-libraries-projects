```python
import tensorflow as tf

# Original tensor
tensor = tf.constant([[4.6, 4.2],
                      [7.5, 3.6],
                      [2.7, 9.4],
                      [6.7, 8.3]], dtype=tf.float32)

# Convert to int
tensor_as_int = tf.cast(tensor, tf.int32)

# Print both
print("Original Tensor:\n", tensor)
print("\nConverted to int:\n", tensor_as_int)
```

```python
a = tf.transpose(tensor_as_int)
a
```

```python
import tensorflow as tf

# Matrix A (2x2)
A = tf.constant([[5, 8],
                 [3, 9]])

# Vector v (2x1)
v = tf.constant([[4],
                 [2]])

# Matrix multiplication
Av = tf.matmul(A, v)

print("Result:\n", Av)
```

```python
# Element-wise multiplication
Av = tf.multiply(A,v)
Av
```

```python
import tensorflow as tf

# Correct matrix (3x2)
A = tf.constant([[4, 9],
                 [5, 6],
                 [1, 8]])

# Get shape
rows, columns = A.shape
print("rows:", rows, "columns:", columns)

# Identity matrix
A_identity = tf.eye(num_rows=rows,
                    num_columns=columns,
                    dtype=tf.int32)

print("\nMatrix A:\n", A)
print("\nIdentity Matrix:\n", A_identity)
```
